```r
#Using the Titanic dataset and Bayes Theorem find out which of the following is more lively to survive:

# a boy in 2nd class
# a women in 3rd class 
# a man in 1st class

#upload a link to your github repository
```

```r
library("XLConnect")
```

```r
df <- readWorksheetFromFile("titanic3.xlsx",
                           sheet=1)
head(df,3)
```

```r
# checking for NA's in each column
colSums(is.na(df))
```

```r
# replacing NA with zeroes
df[is.na(df)]=0
```

```r
# create columns based on conditional(boy, 2nd class, survived) for the BOY cathegory
# eliminated the rows that contain <NAN> in any of the tested fields 

df$B <- ifelse(df$sex == "male" & df$age <= 18, 1,0)
df$B2 <- ifelse(df$sex == "male" & df$age <= 18 & df$pclass == 2, 1,0)
df$B2S <- ifelse(df$sex == "male" & df$age <= 18 & df$pclass == 2  & df$survived == 1, 1,0)

sum(df$B)    # number of boys (males under 18)
sum(df$B2)   # number of boys in the 2nd class
sum(df$B2S)  # number of boys in the 2nd class that survived

head(df,3)
```

```r
# create columns based on conditional(woman, 3rd class, survived)  for the WOMEN cathegory
df$W <- ifelse(df$sex == "female", 1,0)
df$W3 <- ifelse(df$sex == "female" & df$pclass == 3, 1,0)
df$W3S <- ifelse(df$sex == "female" & df$pclass == 3 & df$survived == 1, 1,0)

sum(df$W)    # number of WOMEN
sum(df$W3)    # number of WOMEN in the 3rd class
sum(df$W3S)   # number of WOMEN that survived in the 3rd class
```

```r
# create columns based on conditional(man, 1st class, survived) for the MEN cathegory
df$M <- ifelse(df$sex == "male" & df$age > 18, 1,0)
df$M1 <- ifelse(df$sex == "male" & df$age > 18 & df$pclass == 1, 1,0)
df$M1S <- ifelse(df$sex == "male" & df$age > 18 & df$pclass == 1 & df$survived == 1, 1,0)

sum(df$M)
sum(df$M1)
sum(df$M1S)
```

```r
SurvRate_B2S <-sum(df$B2S)/sum(df$B2)
SurvRate_W3S <-sum(df$W3S)/sum(df$W3)
SurvRate_M1S <-sum(df$M1S)/sum(df$M1)
sprintf(" A boy in 2nd class has a %f survival rate.", SurvRate_B2S )
sprintf(" A women in 3rd class has a %f survival rate.", SurvRate_W3S )
sprintf(" A men in 1st class has a %f survival rate.", SurvRate_M1S )
```
