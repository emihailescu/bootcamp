```python
import pandas as pd

Location = "titanic3.xlsx"
df=pd.read_excel(Location)
df.head()
```

```python
def score_to_numeric(sx):
    if sx=='female':
        return 1
    if sx=="male":
        return 0
```

```python
df['sex']=df['sex'].apply(score_to_numeric)
df.head()
```

```python
df['relatives']=df['sibsp']+df['parch']
df
```
